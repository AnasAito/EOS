```
def int64_to_float (row,col):

        a = row[col].split(",")
        if(len(a)==1): return float(a[0])
        else : return float(a[0]+"."+a[1])

for col in data.columns[11:49]:
  # new col name
  col_name = col+"float"
  data[col_name] = data.apply (lambda row: int64_to_float(row,col), axis=1)


data.columns
```

```
## ONE HOT ENCODER FRO MINR_PASSANT
def fix_10mm (row):

        if row["MINR_PASSANT"]=="10 mm" :
            return "10mm"
        else :  return row["MINR_PASSANT"]


# apply 
data["MINR_PASSANT"] = data.apply (lambda row: fix_10mm(row), axis=1)
```
